```python
def plot_yearly_stat(yearly_stats, xlabels, ylabels, title):
    
    fig, ax = plt.subplots(figsize=(14,5))
    x_ticks = len(years)
    idx = np.arange(x_ticks)
    width = 0.35

    for idx in range(0, yearly_stats.shape[0]):
        ax.plot(xlabels, yearly_stats[idx,:], label=ylabels[idx] )
    
    ax.set_xlabel('years')
    ax.set_ylabel('population')
    ax.legend(loc='best')
    fig.suptitle(title)
    plt.show()
```

```python
mother_status = [HEALTHY, SUSCEPTIBLE, INFECTED, CURED] #, DEAD, MATERNAL_DEATH]
years = [i for i in range(10)]

yearly_stats = np.zeros((len(mother_status), len(years)))
for m_idx, m in enumerate(mother_status):
    for y_idx, y in enumerate(years):
        yearly_stats[m_idx, y_idx] = mother_stats[m][y]

plot_yearly_stat(yearly_stats, years, mother_status, 
                 'Yearly Distribution of mother population')
```
